```python
%%writefile task3a.hql

ADD JAR /opt/cloudera/parcels/CDH/lib/hive/lib/hive-contrib.jar;

USE stackoverflow_;

SELECT 
    users.age,
    SUM(IF (posts.post_type_id=1,1,0)) as cuestion_count,
    SUM(IF (posts.post_type_id=2,1,0)) as answer_count
FROM users
inner JOIN posts
ON posts.owner_user_id = users.id
WHERE age is not null
GROUP BY users.age
ORDER BY users.age;
```
